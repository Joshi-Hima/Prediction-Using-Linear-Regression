# Salary Prediction Using Simple Linear Regression

```python
# Machine Learning - Supervised learning and Unsupervised learning

# Supervised Learning -- Labeled data where the output is already known.
  # Train and Test - 100%(Data) - 80%(Train) - 20%(Test)
  # Linear Regression = y = mx +c
    # Simple linear regression - one feature one target
    # Multiple linear regression - one or more than one feature one target.
    #MSE(Mean Sqaure Error), MAE(Mean Absoulte Error), R squared

# Unsupervised Learning -- Unlabeled data, where the iutput is not predefined.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
salary = pd.read_csv('/content/drive/MyDrive/TechAxis(Data Science)/Salary_dataset.csv')
salary.head()
```

```python
salary.drop(columns = "Unnamed: 0", inplace = True)
salary.columns
```

```python
len(salary)
```

```python
len(salary[salary.duplicated()])
```

```python
salary.isnull().sum()
```

```python
plt.figure(figsize = (6,4))
sns.scatterplot(x = 'YearsExperience', y = 'Salary', data = salary)
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# feature always x and predit garne kura always y

X = salary[['YearsExperience']] #features
Y = salary['Salary'] #target
```

```python
X.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
X_train
```

```python
plt.figure(figsize = (6,4))
plt.scatter(X_train, Y_train, color = 'blue', label = 'Training Data')
plt.scatter(X_test, Y_test, color = 'red', label = 'Testing Data')

plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)

plt.figure(figsize = (6,4))
plt.scatter(X_test, Y_test, color = 'red', label = "Testing Set")
plt.plot(X_test, y_pred, color = 'blue', linewidth = 2, label = 'Regression Line')
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
r2 = r2_score(Y_test, y_pred)
r2

# 0 to 1 ko bich ma vaneko perfect correlation
```

```python
plt.figure(figsize = (6,4))
plt.scatter(x = 'age', y = 'experience', data = df)
plt.title('Scatter plot of age vs experience')
plt.xlabel('Age')
plt.ylabel('Experience')
plt.show()
```

```python
x = df[['age']]
y = df[['experience']]
```

```python
X_train_mul, X_test_mul, Y_train_mul, Y_test_mul = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
model = LinearRegression()
model.fit(X_train_mul, Y_train_mul)
```

```python
Y_pred = model.predict(X_test_mul)

plt.figure(figsize = (6,4))
plt.scatter(X_test_mul, Y_test_mul, color = 'red', label = 'Testing Set')
plt.plot(X_test_mul, Y_pred, color = 'blue', linewidth = 2, label = 'Multiple Regression Line')
plt.title('Age vs Experience')
plt.xlabel('Age')
plt.ylabel('Experience')
plt.legend()
plt.show()
```

```python
acc = r2_score(Y_test_mul, Y_pred)
acc
```
